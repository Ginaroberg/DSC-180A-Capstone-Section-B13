## Prepare data for GP

```python
X = xr.open_mfdataset(['../train_val/inputs_historical.nc', '../train_val/inputs_ssp585.nc']).compute()
X
```

```python
Y = xr.concat([outputs_historical.sel(member=2), outputs_ssp585.sel(member=1)], dim='time').compute()
Y
```

```python
# convert pr to mm/day, default unit is kg/m²/s
Y['pr'] *= 86400
Y['pr90'] *= 86400
```

```python
X['SO2'].mean(['latitude', 'longitude']).plot()
```

```python
from eofs.xarray import Eof
bc_solver = Eof(X['BC'])
bc_solver
```

```python
bc_eofs = bc_solver.eofsAsCorrelation(neofs=5)
bc_pcs = bc_solver.pcs(npcs=5, pcscaling=1)
```

```python
bc_eofs
```

```python
X
```

```python
bc_pcs
```

```python
for mode, eof  in bc_eofs.groupby('mode'):
    # Plot the leading EOF expressed as correlation in the Pacific domain.
    eof.plot(subplot_kws={'projection': ccrs.PlateCarree()},
            cbar_kwargs={'label': '', 'orientation':'horizontal'}, vmin=-1.)
    plt.gca().coastlines()
    plt.gca().set_title(f"EOF{mode} expressed as correlation", fontsize=16)
    plt.show()
```

```python
so2_solver = Eof(X['SO2'])
so2_eofs = so2_solver.eofsAsCorrelation(neofs=5)
so2_pcs = so2_solver.pcs(npcs=5, pcscaling=1)

for (mode, eof), (_, pc) in zip(so2_eofs.groupby('mode'), so2_pcs.groupby('mode')):
    # Plot the leading EOF expressed as correlation in the Pacific domain.
    eof.plot(subplot_kws={'projection': ccrs.PlateCarree()},
             cbar_kwargs={'label': '', 'orientation':'horizontal'}, vmin=-1.)
    plt.gca().coastlines()
    plt.gca().set_title(f"EOF{mode} expressed as correlation", fontsize=16)
    plt.show()
    pc.plot()
    plt.show()
```

```python
bc_df = bc_pcs.to_dataframe().unstack('mode')
bc_df.columns = [f"BC_{i}" for i in range(5)]
bc_df
```

```python
so2_df = so2_pcs.to_dataframe().unstack('mode')
so2_df.columns = [f"SO2_{i}" for i in range(5)]
so2_df
```

```python
# Utilities for normalizing the emissions data
min_co2 = 0.
max_co2 = 2400
def normalize_co2(data):
    return data / max_co2

def un_normalize_co2(data):
    return data * max_co2

min_ch4 = 0.
max_ch4 = 0.6
def normalize_ch4(data):
    return data / max_ch4

def un_normalize_ch4(data):
    return data * max_ch4
```

```python
leading_historical_inputs = pd.DataFrame({
    "CO2": normalize_co2(X["CO2"]).values,
    "CH4": normalize_ch4(X["CH4"]).values
}, X['CO2']['time'].values)
```

```python
leading_historical_inputs
```

```python
leading_historical_inputs=pd.concat([leading_historical_inputs, bc_df, so2_df], axis=1)
leading_historical_inputs
```

## Train GP

```python
from esem import gp_model

tas_gp = gp_model(leading_historical_inputs, Y['tas'])
tas_gp.train()
```

## Predict using test data

### Prepare test data

```python
test_X = xr.open_dataset('../test/inputs_ssp245.nc')
test_Y = xr.open_dataset('../test/outputs_ssp245.nc')
```

```python
test_inputs = pd.DataFrame({
    'CO2': normalize_co2(test_X['CO2'].values),
    'CH4': normalize_ch4(test_X['CH4'].values)
}, index=test_X['CO2']['time'].values)

test_inputs=pd.concat([test_inputs, 
                       bc_solver.projectField(test_X["BC"], neofs=5, eofscaling=1).to_dataframe().unstack('mode').rename(columns={i:f"BC_{i}" for i in range(5)}),
                       so2_solver.projectField(test_X["SO2"], neofs=5, eofscaling=1).to_dataframe().unstack('mode').rename(columns={i:f"_{i}" for i in range(5)}),
                       ], axis=1)

test_inputs
```

### prediction

```python
m_tas, v_tas = tas_gp.predict(test_inputs)
```

```python
with plt.style.context('dark_background'):
    test_Y['tas'].sel(time=2050).mean('member').plot(transform=ccrs.PlateCarree(), subplot_kws={'projection': ccrs.PlateCarree()}, cbar_kwargs={'orientation': 'horizontal', 'label':'Temperature change / K'}, vmin=-2)
    plt.gca().set_title('2050 true tas ssp245')
    plt.gca().coastlines()
```
